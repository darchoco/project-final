# Christian Average Word Count

```python
christian_word_df = df[['genre','lyrics','wordcount']]
christian_word_count = christian_word_df[christian_word_df.genre == 'Christian']
christian_word_count
```

```python
christian_word_count_sum = christian_word_count['wordcount'].sum()
christian_word_count = christian_word_count['wordcount'].count()
christian_average_count_of_words_per_song = christian_word_count_sum/christian_word_count
christian_average_count_of_words_per_song
```

# Average Word Count Per Song by Genre

```python
import matplotlib.pyplot as plt
import numpy as np

averagewordcount = [217.63, 211.26, 697.47, 255.36, 249.64]

# Create an array that contains the frequency count of the top 10 words for the song's genre
x_axis = np.arange(0, len(averagewordcount))
plt.bar(x_axis, averagewordcount, color="r", alpha=0.5, align="center")

tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["Christian", "Country", "Rap", "RnB", "Rock"])

plt.xlim(-0.75, len(x_axis)-0.25)

plt.ylim(0, max(averagewordcount)+100)

plt.title("Average Word Count Per Song by Genre")
plt.xlabel("Genre")
plt.ylabel("Average Word Count")

plt.savefig('static/avgwordcount.svg')
plt.savefig('static/avgwordcount.png') 
```
