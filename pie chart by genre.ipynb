```python
import pandas as pd
```

```python
df = pd.read_csv("lyrics.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
## pulling only the genre and the lyrics from the original dataframe
lyrics_test_df = df[['genre','lyrics']]
lyrics_test_df.head()
```

```python
# Pivoting the table so that genre becomes the row names and lyrics are the columns for it's genre

lyrics_test_df_pivot = lyrics_test_df.pivot(columns='genre', values='lyrics')
lyrics_test_df_pivot.head()
```

```python
# Making a new dataframe with only the Christian genre from our pivoted dataframe above and dropping any null values
lyrics_test_df_christian = lyrics_test_df_pivot[['Christian']].dropna()
lyrics_test_df_christian.head()

## this prints out the whole dataframe so you can view up to 999 rows
# with pd.option_context('display.max_rows', 999):
#     print (lyrics_test_df_christian)
```

```python
# Combining our lists into one big list from the Christian genre list

from Models import worddata
import matplotlib.pyplot as plt


positivecount = 0
negativecount = 0
neutralcount = 0
allcount = 0
for word in lyrics_test_df_christian['Christian']:
    positivecount = positivecount+ worddata.word_calcs(word).positivewordcount()
    negativecount = negativecount+ worddata.word_calcs(word).negativewordcount()
    neutralcount = neutralcount+ worddata.word_calcs(word).neutralwordcount()
    allcount = allcount+worddata.word_calcs(word).nostopwordcount()
    
sizes = [round(positivecount/allcount*100,2),round(negativecount/allcount*100,2),round(neutralcount/allcount*100,2)]
labels = 'positive', 'negative', 'neutral'
fig1, ax1 = plt.subplots()
ax1.pie(sizes,labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax1.set(title='Count of Word Types: Christian')
plt.savefig('static/wordcountchristian.png', dpi = 100)
```
